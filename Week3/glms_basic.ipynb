```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example 1: Logistic Regression
# Simulating data
np.random.seed(42)
n = 100
X = np.random.rand(n)
y = (X > 0.5).astype(int)
# Add a constant for the intercept
X_const = sm.add_constant(X)
print(X.shape, y.shape)
```

```python
# Fit logistic regression
model_logit = sm.GLM(y, X_const, family=sm.families.Binomial())
result_logit = model_logit.fit()
print("Logistic Regression Summary:")
print(result_logit.summary())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Example data
X = np.random.rand(100)  # Replace with your actual data
y = (X > 0.5).astype(int)  # Example target labels, replace with actual labels

# Create and fit the model
model = LogisticRegression()
model.fit(X.reshape(-1, 1), y)

# Generate grid for plotting
X_grid = np.linspace(0, 1, 100).reshape(-1, 1)
pred_prob = model.predict_proba(X_grid)[:, 1]  # Probability of class 1

# Plot Logistic Regression
plt.figure(figsize=(8, 6))
plt.scatter(X, y, label="Observed Data", alpha=0.7)
plt.plot(X_grid, pred_prob, color='red', label="Model Fit (Probability)")
plt.xlabel("X")
plt.ylabel("Probability")
plt.title("Logistic Regression: Observed vs Model Fit")
plt.legend()
plt.grid(True)
plt.show()
```
