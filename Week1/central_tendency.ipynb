### Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Distributions

Normal Distribution

```python
data = np.random.standard_normal(100000) # Normally Distributed Data
sns.histplot(data,kde=True, bins=40, color='blue')
```

Uniform Distribution

```python
data = np.random.random(100000) # Uniform Distributed Data
sns.histplot(data,kde=True, bins=40, color='blue')
```

```python
data = np.random.standard_exponential(10000) # Exponentially Distributed Data
sns.histplot(data,kde=True, bins=40, color='blue')
```

### Creating Sample Dataset for Central Tendency Demonstration

```python
# Sample dataset (You can play around and try different data combinations)

data = [12, 15, 12, 18, 20, 20, 20, 25, 22, 20, 18, 12] # Normal Data
print(data)
# Calculate Mean
mean_value = statistics.mean(data)
print(f"Mean: {mean_value}")

# Calculate Median
median_value = statistics.median(data)
print(f"Median: {median_value}")

# Calculate Mode
try:
    mode_value = statistics.mode(data)
    print(f"Mode: {mode_value}")
except statistics.StatisticsError:
    print("Mode: No unique mode found")

# Optional: Display dataset for context
print("\nDataset:", data)
```

```python
sns.histplot(data,kde=True, bins=40, color='blue')
```

```python
print("\nDataset:", data)

# Visualization
plt.figure(figsize=(10, 6))

# Plot data points
plt.scatter(range(len(data)), data, color='blue', label='Data Points')

# Plot Mean, Median, and Mode
plt.axhline(y=mean_value, color='red', linestyle='--', label=f'Mean: {mean_value}')
plt.axhline(y=median_value, color='green', linestyle='-.', label=f'Median: {median_value}')
if 'mode_value' in locals():
    plt.axhline(y=mode_value, color='purple', linestyle='-', label=f'Mode: {mode_value}')

# Add labels and title
plt.title('Visualization of Mean, Median, and Mode')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
# Measures of Variation
# Calculate Variance
variance_value = statistics.variance(data)
print(f"Variance: {variance_value}")

# Calculate Standard Deviation
std_dev_value = statistics.stdev(data)
print(f"Standard Deviation: {std_dev_value}")

# Calculate Range
range_value = max(data) - min(data)
print(f"Range: {range_value}")

# Optional: Display dataset for context
print("\nDataset:", data)
```

```python
# Visualization
plt.figure(figsize=(12, 8))

# Plot data points
plt.scatter(range(len(data)), data, color='blue', label='Data Points')

# Plot Mean, Median, and Mode
plt.axhline(y=mean_value, color='red', linestyle='--', label=f'Mean: {mean_value}')
plt.axhline(y=median_value, color='green', linestyle='-.', label=f'Median: {median_value}')
if 'mode_value' in locals():
    plt.axhline(y=mode_value, color='purple', linestyle='-', label=f'Mode: {mode_value}')

# Plot Variance as a shaded region
plt.fill_between(range(len(data)), mean_value - std_dev_value, mean_value + std_dev_value, 
                 color='yellow', alpha=0.3, label=f'1 Std Dev (Variance Range)')

# Add labels and title
plt.title('Visualization of Mean, Median, Mode, and Variance')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.grid()

# Show plot
plt.show()
```
